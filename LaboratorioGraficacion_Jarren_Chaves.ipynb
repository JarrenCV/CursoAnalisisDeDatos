```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/casasboston.csv')

datos = pd.DataFrame(df)
print(datos)
```

```python
#df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]] no se necesita para el rename

df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA",
})
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(color="green",orientation="horizontal",histtype="step")

plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=0.2,color="orange")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()

valor_por_ciudad.head(10).plot.barh(color="lightblue")
plt.show()
```
